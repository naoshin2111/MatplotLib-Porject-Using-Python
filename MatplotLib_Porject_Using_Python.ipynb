You have been provided with a dataset that lists Ohio State's leading causes of death from the year 2012.

Using the two data points cause of death and percentile, draw a pie chart to visualize the dataset

```python
#import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data from Ohio state, leading causes of death in the state from year 2012
#cause of death
cause = 'Chronic Diseases', 'Unintentional Injuries', 'Alzheimers', 'Influenza and Pneumonia','Sepsis','Others'
#percentile
percentile = [62,5,4,2,1,26]
```

```python
#set the pie chart plot properties
#set figure size
plt.figure(figsize=(10,10))
#explode the largest pie in the dat set
explode = (0.05,0,0,0,0,0)
#set pie charts properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle = 70)
#set axis equal to draw pie as circle
plt.axis('equal')
#set title of the pie chart
plt.title('Ohio State - 2012 : Leading cause of Death')
#show the plot(pie chart)
plt.show()
```

```python
df = pd.read_csv('ADANIPORTS.csv', parse_dates=True)
```

```python
df.head()
```

```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
#Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
#Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['Volume'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```
